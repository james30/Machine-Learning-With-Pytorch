```python
# Simple PyTorch example
import torch

# create a tensor
x = torch.tensor([[1, 2], [3, 4]])

# move to GPU if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
x = x.to(device)

print('Tensor on device:', device)
print(x)
```
